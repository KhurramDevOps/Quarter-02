```python
from google.colab import userdata
GEMINI_API_KEY= userdata.get('GEMINI_API_KEY')
```

```python
import google.generativeai as genai
```

```python
genai.configure(api_key=GEMINI_API_KEY)
```

```python
model = genai.GenerativeModel("gemini-1.5-flash")
```

```python
import PIL.Image as img
image = img.open("/content/istockphoto-2149530993-1024x1024.jpg")
display(image)
```

```python
promt = "explain the image in detail"
response = model.generate_content([image,promt])
print(response.text)
```

```python
import requests
from io import BytesIO
from PIL import Image
from IPython.display import display

url = "https://img.freepik.com/free-vector/hand-drawn-heart-drawing-illustration_52683-162327.jpg"
response = requests.get(url)
img1 = Image.open(BytesIO(response.content))
display(img1)
```

```python
prompt1 = "explain the image in detail"
```

```python
response1 = model.generate_content([img1,prompt1])
print(response1.text)
```
